## Encoding the categorical variable

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
encoded=labelencoder.fit_transform(salary_data_train['Salary'])
salary=pd.DataFrame(encoded,columns=['salary'])
salary_data_train['salary']=salary
```

```python
salary_data_train
```

```python
salary_data_train.drop('Salary',axis=1,inplace=True)
```

```python
salary_data_train.head()
```

## Encode the input categorical variables using Labelencoding

```python
salary_data_train['education'].unique()
```

```python
salary_data_train['educationno'].unique()
```

## Since education and education no represents the same either one of them can be dropped

```python
salary_data_train.drop('education',axis=1,inplace=True)
```

```python
salary_data_train.head()
```

```python
encoded1=labelencoder.fit_transform(salary_data_train['workclass'])
workclass=pd.DataFrame(encoded1,columns=['Workclass'])
salary_data_train['Workclass']=workclass

encoded2=labelencoder.fit_transform(salary_data_train['maritalstatus'])
ms=pd.DataFrame(encoded2,columns=['ms'])
salary_data_train['MaritalStatus']=ms

encoded3=labelencoder.fit_transform(salary_data_train['occupation'])
o=pd.DataFrame(encoded3,columns=['o'])
salary_data_train['Occupation']=o

encoded4=labelencoder.fit_transform(salary_data_train['relationship'])
r=pd.DataFrame(encoded4,columns=['r'])
salary_data_train['Relationship']=r

encoded5=labelencoder.fit_transform(salary_data_train['race'])
race=pd.DataFrame(encoded5,columns=['race'])
salary_data_train['Race']=race

encoded6=labelencoder.fit_transform(salary_data_train['sex'])
s=pd.DataFrame(encoded6,columns=['s'])
salary_data_train['Sex']=s

encoded7=labelencoder.fit_transform(salary_data_train['native'])
n=pd.DataFrame(encoded7,columns=['n'])
salary_data_train['Native']=n
```

```python
salary_data_train.head()
```

```python
salary_data_train.drop(['workclass','maritalstatus','occupation','relationship','race','sex','native'],axis=1,inplace=True)
```

```python
salary_data_train
```

```python
X_train=salary_data_train.loc[:,salary_data_train.columns != 'salary']
Y_train=salary_data_train[['salary']]
```

```python
X_train.head()
```

```python
Y_train.head()
```

## Scaling down the data

```python
from sklearn.preprocessing import MinMaxScaler
minmaxscalar=MinMaxScaler()
X_train_scaled=minmaxscalar.fit_transform(X_train)
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
```

```python
X_train_scaled.head()
```

# Model Building

```python
from sklearn.svm import SVC
svc=SVC(C=2.0,
    kernel='rbf',
    degree=3,
    gamma='scale')
svc.fit(X_train_scaled,Y_train)
```

```python
Y_pred_train=svc.predict(X_train_scaled)
Y_pred_train
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
training_accuracy_score=accuracy_score(Y_train,Y_pred_train)
training_accuracy_score
```
