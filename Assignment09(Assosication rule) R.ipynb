```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv("C:\\Users\\Rahul\\Downloads\\book.csv")
book
```

```python
df=pd.get_dummies(book)
df
```

```python
frequent_itemsets=apriori(df,min_support=0.4,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets1=apriori(df,min_support=0.2,use_colnames=True)
frequent_itemsets1
```

```python
rules=association_rules(frequent_itemsets1,metric='lift',min_threshold=0.3)
rules
```

```python
rules.sort_values('lift',ascending=False).head(15)
```
